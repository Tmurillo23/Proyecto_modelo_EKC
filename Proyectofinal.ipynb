```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Limpieza de datos


```python
co2_em = pd.read_csv("/content/sample_data/co-emissions-per-capita.csv")
co2_em.pop('Code');
pib = pd.read_csv("/content/sample_data/gdp-per-capita-in-us-dollar-world-bank.csv")
pib.pop('Code');
rnw_energy =pd.read_csv("/content/sample_data/per-capita-renewables.csv")
rnw_energy.pop('Code');
```

```python
# Data Frame Colombia
co2_em_col =co2_em[(co2_em['Entity']=='Colombia') & (co2_em['Year'] >= 1970) & (co2_em.pop('Entity'))]
co2_em_col.pop('Year');
pib_col = pib[(pib['Entity']=='Colombia') & (pib['Year'] >= 1970) & (pib.pop('Entity'))]
pib_col.pop('Year');
rnw_energy_col = rnw_energy[(rnw_energy['Entity']=='Colombia') & (rnw_energy['Year'] >= 1970) & (rnw_energy['Year'] < 2021) & (rnw_energy.pop('Entity'))]
rnw_energy_col.pop('Year');
pib_col_2= pib_col.reset_index()
co2_em_col_2 = co2_em_col.reset_index()
rnw_energy_col_2 = rnw_energy_col.reset_index()
col = pd.concat([co2_em_col_2,rnw_energy_col_2,pib_col_2], axis = 1, ignore_index=True)
col.pop(0)
col.pop(2)
col.pop(4)
col.to_csv('col.csv', index= False, header = False , sep = " ")
y_col = pib_col.to_numpy()
```

```python
# Data Frame Alemania
co2_em_ger =co2_em[(co2_em['Entity']=='Germany') & (co2_em['Year'] >= 1970)  &  (co2_em['Year'] < 2021) & (co2_em.pop('Entity'))]
co2_em_ger.pop('Year');
pib_ger =  pib[(pib['Entity']=='Germany') & (pib['Year'] >= 1970) & (pib.pop('Entity'))]
pib_ger.pop('Year');
rnw_energy_ger = rnw_energy[(rnw_energy['Entity']=='Germany') & (rnw_energy['Year'] >= 1970) & (rnw_energy['Year'] < 2021) & (rnw_energy.pop('Entity'))]
rnw_energy_ger.pop('Year');
pib_ger_2= pib_ger.reset_index()
co2_em_ger_2 = co2_em_ger.reset_index()
rnw_energy_ger_2 = rnw_energy_ger.reset_index()
ger = pd.concat([co2_em_ger_2,rnw_energy_ger_2, pib_ger_2], axis = 1, ignore_index=True)
ger.pop(0)
ger.pop(2)
ger.pop(4)
ger.to_csv('ger.csv', index= False, header = False, sep = " ")
y_ger = pib_ger.to_numpy()
```

# Curvas EKC de cada país

```python
# Alemania,  Alemania si tiene Turning Point
with open(r"/content/resultados_alemania.dat") as ger_dat:
    E0 = [float(data_g.split()[0]) for data_g in ger_dat]
tp_ger = E0[-1]
E0.pop(-1)
E0.pop(0)
plt.plot(y_ger,E0)
plt.title("EKC para Alemania")
plt.xlabel("PIB per cápita en USD($)")
plt.ylabel("Emisiones de C02")
plt.savefig("Gráfica_alemania.jpg")
```
